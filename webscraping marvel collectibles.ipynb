```python
#import the necessary libraries and module
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

title=[]
price=[]
relative_url=[]
#the number of pages is 6
for i in range(1,7):
    url='https://queenstudios.shop/collections/marvel-collectibles?page=' + str(i)
    
    # request to  website
    response=requests.get(url)
    
    # soup object
    soup=BeautifulSoup(response.content,'html.parser')
    results=soup.find_all('div',class_='grid-product__content')
    
    # loop through results
    for result in results:
        title.append(result.find('div',class_='grid-product__title').get_text())
        price.append(result.find('div',class_='grid-product__price').text.strip())
        relative_url.append(result.find('a',class_='grid-product__link').get('href'))
```

```python
root_url='https://queenstudios.shop/collections/marvel-collectibles'
```

```python
# import the module to create the combine_url
import urllib.parse
```

```python
# to get the link of each product_title
# join the relative_url which has the "href" and the root_url


combine_url=[]
for link in relative_url:
    combine_url.append(urllib.parse.urljoin(root_url,link))
```

```python
# create a dataframe for marvel_collectibles and add a dictionary

marvel_collectibles=pd.DataFrame({'Product Title':title,
                                 'Price':price,
                                'Links':combine_url})
```
